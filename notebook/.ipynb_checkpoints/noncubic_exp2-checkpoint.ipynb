```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams.update({'font.size': 12})
```

```python
theta2_graus = [32.62, 34.86, 37.08, 48.34, 63.68, 69.30, 73.18, 82.24, 100.16, 105.24]
theta2 = np.radians(np.array(theta2_graus))
sin2_theta = np.sin(theta2/2)**2

llambda = 1.5405e-10
d = (np.sqrt((llambda**2)/(4*sin2_theta)))
d
```

### Dados de ângulos de pico do Magnésio

## Hexagonal Close Packed - Magnésio

$$
\frac{1}{d^{2}}=\frac{4}{3}\left(\frac{h^{2}+h k+k^{2}}{a^{2}}\right)+\frac{l^{2}}{c^{2}}=\frac{4 \sin ^{2} \theta}{\lambda^{2}}
$$

$$\Downarrow$$

$$
2 \log d=2 \log a-\log \left[\frac{4}{3}\left(h^{2}+h k+k^{2}\right)+\frac{l^{2}}{(c / a)^{2}}\right]
$$

```python
def s_magnesio(h, k, l, c_a):
    return (4/3*(h**2+h*k+k**2)+(l**2)/(c_a**2))

def s_tetragonal(h, k, l, c_a):
    return ((h**2+k**2)+(l**2)/(c_a**2))
```

```python
c_a = np.linspace(0.5,2, 500)
hkl = [(1,0,0),(1,0,1),(1,0,2),(1,1,0), (0,0,2),
      (1,0,3),(2,0,0),(1,1,2),(2,0,1),(0,0,4),
      (2,0,2),(1,0,4),(2,0,3),(2,1,0)]




fig, ax0 = plt.subplots(1, 2, figsize=(16, 8))

cmap = cm.get_cmap('gist_rainbow')
size = len(hkl)

colors = [cmap(i/size) for i in range(size)]

ax = ax0[0]
i=0
for aux in hkl:
    h, k, l = aux
    y = s_magnesio(h, k, l, c_a)
    ax.semilogy(c_a, y, color=colors[i], label = 'hkl = '+str(aux[0])+str(aux[1])+str(aux[2]))
    i+=1
i=0

ax.set_xticks(np.arange(min(c_a), max(c_a)+0.1, 0.1))
ax.xaxis.set_minor_locator(plt.MultipleLocator(0.1 / 10))

ax.set_ylim(0.5, 10)
ax.grid(True)
ax.invert_yaxis()
ax.set_ylabel('log(s)', size=18)
ax.set_xlabel('(c/a)', size=18)

tax = ax0[1]

for aux in sin2_theta:
    tax.axhline(aux, color = 'k', linestyle = '-', linewidth = 2)
    
tax.set_yscale('log')
tax.set_ylim(0.05, 1)
tax.invert_yaxis()
tax.set_ylabel('$\sin^2(\\theta)$', size=18)

ax.legend(bbox_to_anchor=(1.07, 0.5), loc="upper left", borderaxespad=0)



plt.tight_layout()

plt.savefig('hull-davey_completo.svg', dpi=300)
plt.savefig('hull-davey_completo.png', dpi=300)
```

## Tetragonal - Estanho

$$
\frac{1}{d^2}=\frac{1}{a^2}\left( (h^2+k^2) + \frac{l^2}{(c/a)^2} \right) = \frac{4 \sin ^{2} \theta}{\lambda^{2}}
$$

$$\Downarrow$$

$$
2 \log d=2 \log a-\log \left( (h^2+k^2) + \frac{l^2}{(c/a)^2} \right)
$$
